```python
import re
files = []
result = 0

pattern = re.compile(r'(\$ cd (.*)\n\$ ls\n(.[^(\$)]+))') # create regex pattern that finds each section of the input that lists the contents of a folder
input = list(filter(None, pattern.split(open('input.txt').read()))) # create list from input, with sections listing the files broken up by folder

for section in input: # go through each section of the input
    folder = pattern.search(section) # use regex pattern to extract the different sections out of the text
    if folder: # make sure the regex pattern isn't empty
        folder_name = folder.group(2) # extract the name of the folder
        folder_content = folder.group(3).splitlines() # extract the content of the folder
        added_folder = [folder_name] # create a list starting with the folder name
        for content in folder_content: # go through the contents of the folder
            added_folder.append(content.split()[1]) # add each item inside the folder to the list
            if re.findall(r'[1-9]', content): # only if it's a file:
                files.append([content.split()[1], int(content.split()[0])]) # add the file and the size of the file as an int.
        files.append(added_folder) # add the the entry to the greater list of files inside the input

def calculate_size(folder): # calculate the size of the given folder
    size = 0
    for item in folder[1:]: # go through all the lists in files # go through each item in the list, skipping the first one
        if isinstance(folder[1], int): # check if the item is a file
            size += item # add the size of the file to the count
            #print("added " + str(item) + " from " + str(folder[0]))
        elif isinstance(folder[1], str): # check if the item is a folder
            for entry in files: # go through all the files again
                if item == entry[0]: # find the list that corresponds to the item found
                    size += calculate_size(entry) # do this function again
    return size

for folder in files:
    if calculate_size(folder) <= 100000 and isinstance(folder[1], str): # find the folders that are smaller than 100000 check it's a folder by making sure it doesn't contain a number
        result += calculate_size(folder)

print(result)
```
